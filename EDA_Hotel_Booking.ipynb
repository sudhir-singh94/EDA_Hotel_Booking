# **Project Summary -**

In this project, we explored a dataset containing hotel booking information for a city hotel and a resort hotel. This dataset provided valuable insights about the patterns and factors that govern hotel bookings, enabling us to answer questions about the best time of year to book a hotel room, the optimal length of stay for the best daily rate, and the likelihood of a hotel receiving a disproportionately high number of special requests.

The dataset included a variety of information such as when the booking was made, the length of stay, the number of adults, children, and babies, and the number of available parking spaces, among other things. Importantly, all personally identifying information had been removed from the data, ensuring the privacy of the guests while still allowing for rich analysis.

Through our exploration and analysis, we found that the city hotel had more bookings compared to the resort hotel, with the busiest months for bookings being August, July, and May. This suggests that the peak season for both city and resort hotels is the summer, possibly due to the holidays during this period. It was also found that the most common type of customer making bookings is transient, suggesting that these hotels are often used for temporary stays, such as business trips or short vacations.

In terms of booking behavior, we found that more bookings are not canceled than canceled, which is a positive sign for hotel occupancy. However, bookings made further in advance were found to be more likely to be canceled, possibly due to changes in plans or circumstances. Additionally, bookings with changes were less likely to be canceled, perhaps indicating that customers who are engaged enough to make changes to their booking are less likely to cancel altogether.

We also investigated special requests and found that the city hotel tends to receive more special requests than the resort hotel. Among customer types, transient-party customers were found to make the most special requests, which could be due to the nature of these bookings, which often involve larger groups with more complex needs.

Interestingly, we also found correlations between various booking characteristics. For example, a longer total stay was associated with more weekend nights and weeknight stays. Also, customers who had more previous cancellations were more likely to be repeated guests, perhaps indicating a stronger relationship with the hotel despite the cancellations.

From a revenue perspective, we found that the city hotel makes significantly more total revenue than the resort hotel, likely due to its higher number of bookings. Furthermore, we discovered that the average daily rate is higher in the summer months, suggesting that the demand for hotel bookings is seasonal with higher prices during the peak season.

In conclusion, this project provided valuable insights into the patterns and factors that govern hotel bookings. Such insights can help hoteliers make informed decisions about pricing, marketing, and service offerings. Additionally, understanding these patterns can help customers make better booking decisions, optimizing their stay based on their personal needs and preferences. However, it's important to note that these insights are based on the specific dataset we analyzed, and may not generalize to all hotels or regions.

# **Problem Statement**

The hospitality industry heavily depends on understanding customer behavior to optimize booking rates, maximize revenue, and improve customer satisfaction. Despite the vast amounts of data generated by hotel bookings, many hoteliers lack the necessary data analysis capabilities to harness these data for actionable insights. Therefore, the problem is to explore and analyze a hotel booking dataset to uncover the key patterns and factors that govern hotel bookings. Specifically, the analysis aims to determine the optimal time of year to book a hotel room, the ideal length of stay for the best daily rate, and the factors affecting the number of special requests. The findings from this analysis could help hoteliers make informed business decisions about pricing, marketing, and service strategies, and customers make better booking decisions, optimizing their stay based on their personal needs and preferences.

**Define Your Business Objective?**

1. **Maximize Occupancy Rates:** By understanding patterns in booking behavior, hotels can adjust their strategies to ensure maximum occupancy throughout the year. This could involve strategic pricing during off-peak periods or targeted marketing to attract guests during typically quiet periods.

2. **Optimize Pricing Strategy:** With insights into when and how customers book rooms, hotels can better tailor their pricing strategies. This could mean higher prices during peak booking times or offering discounts for longer stays.

3. **Enhance Customer Satisfaction:** By predicting the likelihood of special requests, hotels can better prepare for guest needs, leading to improved guest satisfaction and potentially more positive reviews and repeat customers.

4. **Efficient Resource Allocation:** Understanding the factors that influence cancellations can help hotels better manage room inventory and reduce instances of overbooking or underbooking.

5. **Tailored Marketing Efforts:** Insights into customer behavior can guide more effective marketing efforts. For instance, knowing the most popular booking channels can help hotels invest wisely in their marketing strategies.

6. **Increase Revenue:** All of the above objectives ultimately aim to increase hotel revenue. Whether it's through increased occupancy, optimized pricing, reduced cancellations, or effective marketing, the end goal is to improve the hotel's bottom line.

2.3 Handling missing Data

The dataset has missing values in the following columns:

children: 4 missing values

country: 488 missing values

agent: 16,340 missing values

company: 112,593 missing values

Let's handle these missing values. The way we handle them depends on the nature of the data:

For children, since there are only a few missing values, we could fill them with the mode (most common) value.

For country, we could fill the missing values with the mode (most common country).

For agent and company, these are identifiers for the agent and company responsible for the booking. It could be the case that NaN values indicate that the booking was made without an agent or a company, so we can fill these with a new category, like 'None'.

```python
# Fill missing values

# For 'children' and 'country', fill with the most frequent value (mode)
df['children'].fillna(df['children'].mode()[0], inplace=True)
df['country'].fillna(df['country'].mode()[0], inplace=True)

# For 'agent' and 'company', fill with a new category 'None'
df['agent'].fillna('None', inplace=True)
df['company'].fillna('None', inplace=True)

# Check if there are any missing values left
df.isnull().sum().sum()
```

## 13. Plot the average daily rate by month

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already loaded the DataFrame as 'df'

# Create a new column 'arrival_date_month_num' to represent months in numerical form for easy sorting
df['arrival_date_month_num'] = pd.to_datetime(df['arrival_date_month'], format='%B').dt.month

# Group by 'arrival_date_month_num' and calculate the mean 'adr'
avg_rate_by_month = df.groupby('arrival_date_month_num')['adr'].mean().reset_index()

# Plot the average daily rate by month
plt.figure(figsize=(8, 6))
sns.lineplot(data=avg_rate_by_month, x='arrival_date_month_num', y='adr')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title('Average Daily Rate by Month')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate')
plt.show()
```

The line plot of average daily rate by month shows that the rates are higher in the summer months (June, July, August) and lower in the winter months (November, December, January). This suggests that demand for hotel bookings is seasonal, with higher demand (and therefore higher prices) during the summer.

**Special Requests**: The number of special requests made by the customer influences the booking and cancellation rates. Generally, fewer special requests are associated with a higher cancellation rate. This suggests that fulfilling special requests could improve guest satisfaction and reduce cancellations.

**Assigned Room**: The cancellation rate is higher when the same room type as demanded was not allotted. Ensuring that customers receive their desired room type could help reduce cancellations.

**Distribution Channel**: The Travel Agents/Operators distribution channel brings in the highest average revenue for the hotels, but also has the highest cancellation rate. This implies a need to manage relationships with these agents effectively to maximize revenue and minimize cancellacies.


Overall, there are several factors that influence booking and cancellation rates, customer return rates, and revenue. By understanding these factors and tailoring strategies accordingly, the hotels can improve customer satisfaction, reduce cancellations, and ultimately increase their revenue.
